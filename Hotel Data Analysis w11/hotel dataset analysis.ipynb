```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_excel('/content/hotel_dataset.xlsx')
review = pd.DataFrame(data)
```

```python
review.describe()
```

```python
review.info()
```

```python
avg_spending_by_city = review.groupby('yasadigi_sehir')['toplam_harcama'].mean().sort_values(ascending=False)
display(avg_spending_by_city.head(10))
```

```python
data_selected = review[['ortalama_rezervasyon_oncesi_gun', 'toplam_harcama']]
corr_m = data_selected.corr(method='pearson')
display(corr_m)
correlation_coefficient = corr_m.loc['ortalama_rezervasyon_oncesi_gun', 'toplam_harcama']

print(f"ortalama_rezervasyon_oncesi_gun ve toplam_harcama arasındaki Korelasyon Katsayısı (r): {correlation_coefficient:.4f}")
```

```python
df_selected = review[['toplam_harcama', 'ziyaret_amaci']]

plt.style.use('seaborn-v0_8-whitegrid')

plt.figure(figsize=(10, 6))

sns.boxplot(x='ziyaret_amaci', y='toplam_harcama', data=df_selected, palette='pastel')

plt.title('Ziyaret Amacına Göre Toplam Harcama Dağılımı', fontsize=16)
plt.xlabel('Ziyaret Amacı', fontsize=14)
plt.ylabel('Toplam Harcama', fontsize=14)

plt.ticklabel_format(style='plain', axis='y')

plt.show()
```
