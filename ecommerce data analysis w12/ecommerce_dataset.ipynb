```python
df_clean = df.copy()

# time_on_site_sec - fill with median by device
if 'time_on_site_sec' in df_clean.columns:
  device_median = df_clean.groupby('device')['time_on_site_sec'].transform('median')
  df_clean['time_on_site_sec'] = df_clean['time_on_site_sec'].fillna(device_median)
  df_clean['time_on_site_sec'] = df_clean['time_on_site_sec'].fillna(df_clean['time_on_site_sec'].median())

# 2. hour - fill with mode or -1
if 'hour' in df_clean.columns:
  df_clean['hour'] = pd.to_numeric(df_clean['hour'], errors='coerce')
  mode_hour = df_clean['hour'].mode()[0] if not df_clean['hour'].mode().empty else -1
  df_clean['hour'] = df_clean['hour'].fillna(mode_hour)

# 3. discount_rate - fill based on discount_used
if 'discount_rate' in df_clean.columns and 'discount_used' in df_clean.columns:
  mask = (df_clean['discount_used'] == 0) & (df_clean['discount_rate'].isna())
  df_clean.loc[mask, 'discount_rate'] = 0

  category_median = df_clean.groupby('product_category')['discount_rate'].transform('median')
  df_clean['discount_rate'] = df_clean['discount_rate'].fillna(category_median)
  df_clean['discount_rate'] = df_clean['discount_rate'].fillna(df_clean['discount_rate'].median())

# 4. payment_risk_score - fill with mean by user_segment
if 'payment_risk_score' in df_clean.columns and 'user_segment' in df_clean.columns:
  segment_mean = df_clean.groupby('user_segment')['payment_risk_score'].transform('mean')
  df_clean['payment_risk_score'] = df_clean['payment_risk_score'].fillna(segment_mean)
  df_clean['payment_risk_score'] = df_clean['payment_risk_score'].fillna(df_clean['payment_risk_score'].mean())

# 5. Columns like returned, refund_amount, delivery_days - fill with 0 for non-purchases

zero_fill_score = ['returned', 'refund_amount', 'delivery_days', 'delivery_delay']
for col in zero_fill_score:
  if col in df_clean.columns:
    df_clean[col] = pd.to_numeric(df_clean[col], errors='coerce')
    df_clean[col] = df_clean[col].fillna(0)
```

```python
# Handle the Remaining Missing Values

df_clean['has_payment_method'] = df_clean['payment_method'].notna().astype(int)
df_clean['has_order_id'] = df_clean['order_id'].notna().astype(int)

df_clean['payment_method'] = df_clean['payment_method'].fillna('no_payment')
df_clean['order_id'] = df_clean['order_id'].fillna('no_order')

categorical_fill_cols = ['device', 'traffic_source', 'city', 'user_segment']

for col in categorical_fill_cols:
    if col in df_clean.columns:
        # Fill with 'unknown' or mode
        mode_val = df_clean[col].mode()[0] if not df_clean[col].mode().empty else 'unknown'
        df_clean[col] = df_clean[col].fillna(mode_val)
        print(f"Filled {col} missing with: {mode_val}")
```

```python
# 1. Engagement features
df_clean['views_per_second'] = df_clean['page_views'] / (df_clean['time_on_site_sec'] + 1)
df_clean['engagement_score'] = df_clean['page_views'] * df_clean['session_quality_score']

# 2. Time features (if date_text exists)
if 'date_text' in df_clean.columns:
    # Convert to datetime
    df_clean['date'] = pd.to_datetime(df_clean['date_text'], format='%d.%m.%Y', errors='coerce')
    df_clean['month'] = df_clean['date'].dt.month
    df_clean['day_of_week'] = df_clean['date'].dt.dayofweek
    df_clean['is_month_end'] = df_clean['date'].dt.is_month_end.astype(int)
    df_clean['is_month_start'] = df_clean['date'].dt.is_month_start.astype(int)

# 3. Discount impact
df_clean['discount_impact'] = df_clean['discount_rate'] * df_clean['discount_used']

# 4. Risk features
df_clean['high_risk'] = (df_clean['payment_risk_score'] > 0.7).astype(int)

# 5. Session intensity
df_clean['intensive_session'] = ((df_clean['page_views'] > df_clean['page_views'].median()) &
                                  (df_clean['time_on_site_sec'] > df_clean['time_on_site_sec'].median())).astype(int)

print("New features created:")
new_features = [col for col in df_clean.columns if col not in df.columns]
print(new_features)
```

# MODEL1

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, mean_squared_error, r2_score
import xgboost as xgb
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
import warnings
warnings.filterwarnings('ignore')

df = pd.read_csv('/content/ecommerce_cleaned_ready.csv')
```

```python
# QUESTION 1: WILL CUSTOMER BUY AGAIN? (Classification)

X1 = df.drop(['returning_user', 'session_id', 'user_id', 'data_text', 'order_id', 'payment_method', 'converted', 'purchase_amount', 'date'], axis=1, errors='ignore')
y1 = df['returning_user']

categorical_cols = X1.select_dtypes(include=['object']).columns.tolist()
numerical_cols = X1.select_dtypes(include=['int64', 'float64']).columns.tolist()

X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y1, test_size=0.2, random_state=42, stratify=y1)

preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_cols),
        ('cat', OneHotEncoder(handle_unknown='ignore', sparse_output=False), categorical_cols)
    ])

model1 = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', RandomForestClassifier(
        n_estimators=100,
        max_depth=10,
        random_state=42,
        class_weight='balanced'
    ))
])

model1.fit(X1_train, y1_train)

y1_pred = model1.predict(X1_test)
y1_pred_proba = model1.predict_proba(X1_test)[:, 1]

print(f"Accuracy: {accuracy_score(y1_test, y1_pred):.4f}")

print('\nClassification Report')
print(classification_report(y1_test, y1_pred))

print('\nConfusion Matrix')
print(confusion_matrix(y1_test, y1_pred))
```

# MODEL2

```python
# QUESTION 2: HOW MUCH WILL CLIENT SPEND? (Regression)

df_converted = df[df['converted'] == 1].copy()

X2 = df_converted.drop(['purchase_amount', 'session_id', 'user_id', 'date_text', 'order_id', 'payment_method', 'converted', 'returning_user', 'date'], axis=1, errors='ignore')
y2 = df_converted['purchase_amount']

X2_train, X2_test, y2_train, y2_test = train_test_split(X2, y2, test_size=0.2, random_state=42)

categorical_cols2 = X2.select_dtypes(include=['object']).columns.tolist()
numerical_cols2 = X2.select_dtypes(include=['int64', 'float64']).columns.tolist()

preprocessor2 = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_cols2),
        ('cat', OneHotEncoder(handle_unknown='ignore', sparse_output=False), categorical_cols2)
    ]
)

model2 = Pipeline(steps=[
    ('preprocessor', preprocessor2),
    ('regressor', RandomForestRegressor(
        n_estimators=100,
        max_depth=10,
        random_state=42
    ))
])

model2.fit(X2_train, y2_train)

y2_pred = model2.predict(X2_test)

print('\nModel2 performance:')
print(f'RMSE (Root Mean Squared Error): {np.sqrt(mean_squared_error(y2_test, y2_pred)):.2f}')
print(f'MAE (Mean Absolute Error): {np.mean(np.abs(y2_test - y2_pred)):.2f}')
print(f'R2 Score: {r2_score(y2_test, y2_pred):.2f}')

results_df = pd.DataFrame({
    'Actual': y2_test.values[:10],
    'Predicted': y2_pred[:10],
    'Difference': y2_test.values[:10] - y2_pred[:10]
})

print('\nFirst 10 Predictions:')
print(results_df)
```

```python
# FEATURE IMPORTANCE

# Get feature names after one-hot encoding
preprocessor.fit(X1_train)
feature_names = []
for name, trans, cols in preprocessor.transformers_:
    if name == 'cat':
        # Get one-hot encoded feature names
        ohe = trans
        feature_names.extend(ohe.get_feature_names_out(cols))
    elif name == 'num':
        feature_names.extend(cols)

# Get feature importances
importances = model1.named_steps['classifier'].feature_importances_

# Create importance DataFrame
importance_df = pd.DataFrame({
    'Feature': feature_names,
    'Importance': importances
}).sort_values('Importance', ascending=False)

print("Top 20 most important features for predicting returning_user:")
print(importance_df.head(20).to_string())
```
